```python
#Criar função César com chave variável.
from cryptography.fernet import Fernet

def cifra_cesar(texto, chave):
    resultado = ""
    for letra in texto:
        if letra.isalpha():
            desloc = 65 if letra.isupper() else 97
            resultado += chr((ord(letra) - desloc + chave) % 26 + desloc)
        else:
            resultado += letra
    return resultado

chave = Fernet.generate_key()
f = Fernet(chave)

mensagem = "Oi"

cripto = f.encrypt(mensagem.encode())

descripto = f.decrypt(cripto).decode()
print("Mensagem Descriptografada", descripto)


```
